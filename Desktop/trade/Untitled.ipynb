```python
#!/usr/bin/python
# -*- coding: UTF-8 -*-

import statsmodels
import requests
from io import StringIO
import pandas as pd
import numpy as np
import vis
from bs4 import BeautifulSoup


datestr = '20210630'

# 下載股價
r = requests.post('https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + datestr + '&type=ALL')
    
    
# 整理資料，變成表格
df = pd.read_csv(StringIO(r.text.replace("=", "")), 
            header=["證券代號" in l for l in r.text.split("\n")].index(True)-1)

# 整理一些字串：
df_str = df.apply(lambda s: pd.to_numeric(s.astype(str).str.replace(",", "").replace("+", "1").replace("-", "-1"), errors='coerce'))

# 顯示出來
print(len(df_str))
df_str.head().size
```
