```python


def getdict_url(url,txt_dict,i):
    response=requests.get(url)
    bs = BeautifulSoup(response.text, "html.parser")
    for a_tags in bs.find_all('p'):
        str1 = a_tags.text
        str2 = re.sub('^[0-9]+[. |、]','',str1)
        print(str2)
        txt_dict[str2] = i
        i = i + 1
    return txt_dict, i


```
